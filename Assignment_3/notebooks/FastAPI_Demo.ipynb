## Fast API Demonstration


```python
import requests
import zipfile
import io
import os
from PIL import Image
```

Provide the url with the endpoint
and the json file to requests

```python
url = 'http://127.0.0.1:8000/get_predictions_json' 
headers = {'location': 'Nebraska'}
```

```python
#passing the parameters to the request
r = requests.post(url, json = headers,stream=True)
```

# Iteration 1

```python
#Checking if response returns an error
string = str(r.content)
if "Error" not in string:
    
    #Extracting the zip file from response
    with open("output.zip", 'wb') as f:
        f.write(r.content)
    
    #Extracting the zip contents to local
    filepath = "./output"
    with zipfile.ZipFile('output.zip', 'r') as zipObj:
       # Extract all the contents of zip file in different directory
       zipObj.extractall(filepath)
    
    #Displaying the images in the extracted folder
    for file in os.listdir(filepath + "/images/"):
        if file.endswith(".jpg"):
            print(filepath + "images/" + file)
            img = Image.open(filepath + "/images/" + file)
            display(img)
else:
    print(string)
```

# Iteration 2

```python
url = 'http://127.0.0.1:8000/get_predictions_json' 
headers = {'location': 'Boston'}

r = requests.post(url, json = headers,stream=True)

string = str(r.content)
if "Error" not in string:
    with open("output.zip", 'wb') as f:
        f.write(r.content)
    
    filepath = "./output"
    with zipfile.ZipFile('output.zip', 'r') as zipObj:
       # Extract all the contents of zip file in different directory
       zipObj.extractall(filepath)
        
    for file in os.listdir(filepath + "/images/"):
        if file.endswith(".jpg"):
            print(filepath + "images/" + file)
            img = Image.open(filepath + "/images/" + file)
            display(img)
else:
    print(string)
```

# Iteration 3

```python
url = 'http://127.0.0.1:8000/get_predictions_json' 
headers = {
  "location": "South Dakota",
  "state": "string",
  "starttime": "2022-03-11T03:04:46.454Z",
  "endtime": "2022-03-11T03:04:46.454Z"
}

r = requests.post(url, json = headers,stream=True)

string = str(r.content)
if "Error" not in string:
    with open("output.zip", 'wb') as f:
        f.write(r.content)
    
    filepath = "./output"
    with zipfile.ZipFile('output.zip', 'r') as zipObj:
       # Extract all the contents of zip file in different directory
       zipObj.extractall(filepath)
        
    files = os.listdir(filepath + "/images/")

    for file in files:
        if file.endswith(".jpg"):
            print(filepath + "images/" + file)
            img = Image.open(filepath + "/images/" + file)
            display(img)
else:
    print(string)
```
